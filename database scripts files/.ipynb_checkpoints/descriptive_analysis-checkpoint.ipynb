```python
import os.path, sys

import pandas as pd
import numpy as np
import matplotlib as plt

#cards
cards_train = pd.read_csv(os.path.abspath('./../data/train_sets/card_train.csv'), sep=";", dtype='unicode')
cards_test = pd.read_csv(os.path.abspath('./../data/test_sets/card_test.csv'), sep=";", dtype='unicode')

#loans
loans_train = pd.read_csv(os.path.abspath('./../data/train_sets/loan_train.csv'), sep=";", dtype='unicode')
loans_test = pd.read_csv(os.path.abspath('./../data/test_sets/loan_test.csv'), sep=";", dtype='unicode')

#trans
trans_train = pd.read_csv(os.path.abspath('./../data/train_sets/trans_train.csv'), sep=";", dtype='unicode')
trans_test = pd.read_csv(os.path.abspath('./../data/test_sets/trans_test.csv'), sep=";", dtype='unicode')

#unique data
clients = pd.read_csv(os.path.abspath('./../data/client.csv'), sep=";", dtype='unicode')
disp = pd.read_csv(os.path.abspath('./../data/disp.csv'), sep=";", dtype='unicode')
district = pd.read_csv(os.path.abspath('./../data/district.csv'), sep=";", dtype='unicode')
accounts = pd.read_csv(os.path.abspath('./../data/account.csv'), sep=";", dtype='unicode')
```

```python
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
from datetime import datetime

converted_dates = []

print("Month N: " + month + " Month S: " + months[int(float(month)) - 1])

for row in loans_train['date']:
    year = row[:2]
    month = row[2:4]
    day = row[4:6]
    date = "" + day + " " + month + " 19" + year
    converted_dates.append(datetime.strptime(date, "%d %m %Y").strftime('%a, %b %d, \'%y'))
    
loans_train['date'] = converted_dates
loans_train.to_csv(os.path.abspath('./../data/train_sets/loan_train_modified.csv'), sep=";")
```
