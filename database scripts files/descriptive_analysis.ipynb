```python
from datetime import datetime

genders = []
birth_dates = []

for row in clients['birth_number']:
    year = row[:2]
    month = int(float(row[2:4]))
    day = row[4:6]
    if month > 50:
        genders.append('female')
        month -= 50
    else:
        genders.append('male')
    date = "" + day + " " + str(month) + " 19" + year
    birth_dates.append(datetime.strptime(date, "%d %m %Y").strftime('%a, %b %d, \'%y'))
    
clients['birth_date'] = birth_dates
clients['gender'] = genders
del clients['birth_number']

```
